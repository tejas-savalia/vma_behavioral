```python
from scipy.stats import vonmises
import matplotlib.pyplot as plt
import numpy as np
```

```python
kappa = 3.99
mean, var, skew, kurt = vonmises.stats(kappa, moments='mvsk')
```

```python
fig, ax = plt.subplots(1, 1)
```

```python
x = np.linspace(vonmises.ppf(0.01, kappa), vonmises.ppf(0.9999, kappa), 100)
plt.plot(x, vonmises.pdf(x, kappa),'r-', lw=5, alpha=0.6, label='vonmises pdf')
```
